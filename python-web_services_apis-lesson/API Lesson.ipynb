```python
codes_url = 'https://naics-codes.firebaseio.com/codes.json'
res = requests.get(codes_url)
```

```python
codes_dict = res.json()
```

```python
for key, value in codes_dict.items():
    print(key)
    print(value)
    print('')
```

```python
codes = [{'code': code, 'description': description} for code, description in codes_dict.items()]
pd.DataFrame(codes)
```

```python
codes = []
for k, v in codes_dict.items():
    code = {}
    code['code'] = k
    code['description'] = v
    codes.append(code)
codes_df = pd.DataFrame(codes)
```

```python
codes_df.head()
```

```python
weather_key = '661a9cbc625e1aabfbdb9d118cfb156b'
```

```python

forecast_url = 'https://api.openweathermap.org/data/2.5/weather'
zip_code = input('What is your zip?')
params = {
    'APPID': weather_key,
    'zip': '{},us'.format(zip_code),
    'units': 'imperial'
}

res = requests.get(forecast_url, params)

weather_dict = res.json()
weather_dict['main']['temp']
```

```python
url = 'https://api.openweathermap.org/data/2.5/forecast?APPID={}&zip=78723,us&units=imperial'.format(weather_key)
res = requests.get(url)

forecast_dict = res.json()
```
